```python
# Fetch a Wikipedia article
def fetch_wikipedia_content(title, word_count=500):
    page = wikipedia.page(title)
    content = page.content
    words = word_tokenize(content)
    return ' '.join(words[:word_count])
```

**Difference between Penn Treebank and Universal Dependencies Tag-set**
1.  Penn Treebank is more granular and specific, capturing even the smallest distinctions in English grammar. I've added the list of pos-tags supported by Penn Treebank (36) and Universal Dependencies (17) below.
2.  Whereas UD tagset is much more general than Penn Treebank, which makes it usable for other languages too.
3.  UD uses only one verb tag for all whereas PEnn treebank uses tags for different verb forms (base form, past tense, past participle, etc.).
4.  even though Penn Treebank is more nuanced than Universal dependencies, there's no POS-tag for Auxillary Verb in PTB Tag set.

Homographs : words that are spelled the same but have different meanings and sometimes different pronunciations.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import pos_tag
text = "Arun decided to lead the team to victory. The pipes are made of lead"
# Split text into sentences
sentences = sent_tokenize(text)
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}:")
    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
    print(tagged)
    print()
```

If you ntice the word 'lead', you'll notice that in the first sentence it is tagged as a Verb since the word 'lead' is used to describe the action that Arun did and whereas in the second senetence we are talking about the metal 'lead' so it's a noun here.

```python
text=" We need to arrange the chairs in a row. They went out to row on the lake."
sentences = sent_tokenize(text)
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}:")
    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
    print(tagged)
    print()
```

In the 1st sentence, "row" refers to a line of objects or people arranged in a sequence. so, it’s a noun.
whereas in the second senetence, "row" refers to the action of propelling a boat using oars. In this context, it’s a verb meaning to use oars to move a boat across water.

```python
text=" He wound the clock carefully. She wound up from the argument. Alen fell down and got a wound"
sentences = sent_tokenize(text)
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}:")
    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
    print(tagged)
    print()
```

In the 1st sentence,  "wound" is the past tense of the verb "wind," meaning to turn or twist something. In this context, "He wound the clock" means he turned the clock to make it work.
In the 2nd sentence,  "wound" is used as the past participle form of the verb "wind" but is functioning in a different context.
In the 3rd sentence, "wound" is a noun denoting a physical harm resulting from an accident or impact.

The Penn Treebank POS tag-set ensures that there is no ambiguity between the homographs
