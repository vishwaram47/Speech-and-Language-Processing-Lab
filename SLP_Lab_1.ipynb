##Differences between the 3 tokenizers

Word_tokenizer and Wordpunct_tokenizer work almost the same way, the tokens are the words itself but when it comes to wordpunct_tokenuzer they use Regular Expression to split the text on punctuations and whitespaces.

But when it comes to LegalitySyllableTokenizer, the text is split using the LegalityPrinciple and returns a list of lists where each list inside the main list contains the syllabified version of the words.

For example, the word 'commonly' is syllabified to ['com','mon','ly'].

#Stemming

```python
first_2000_words
```

```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
# We can use Wordpunct_tokenizer tokens to do stemming as it splits the text on the basis of punctuations and punctuations shouldn't be there when stemming
words = [word for word in wordpunct_tokens if word.isalnum()]
porter = PorterStemmer()
lancaster = LancasterStemmer()

# Apply Porter Stemmer
porter_stems = [porter.stem(word) for word in words]

# Apply Lancaster Stemmer
lancaster_stems = [lancaster.stem(word) for word in words]
```

```python
print("Original Words:", words)
print("Porter Stemmer Results:", porter_stems)
print("Lancaster Stemmer Results:", lancaster_stems)
```

The difference that I could find is,
Lancaster Stemmer is much more stronger than the Porter Stemmer and reduces the words to it's shortest possible level and doesn't really care about the readability. Eg. for the word 'freely' Porter stems it to 'freeli' but Lancaster stems it to 'fre'.



But still Lancaster maybe be more appropriate in situations where the short words are preferred.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(first_2000_words)
spacy_lemmatized_text = " ".join([token.lemma_ for token in doc])
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

lemmatizer = WordNetLemmatizer()

# Function to map NLTK POS tags to WordNet POS tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

tokens = word_tokenize(first_2000_words)
pos_tags = nltk.pos_tag(tokens)
nltk_lemmatized_text = " ".join([lemmatizer.lemmatize(token, get_wordnet_pos(tag)) for token, tag in pos_tags])
```

```python
def print_side_by_side(original, spacy_lemmas, nltk_lemmas):
    print(f"{'Original Text':<20} | {'SpaCy Lemmatized':<20} | {'NLTK Lemmatized':<20}")
    print("-" * 60)
    original_tokens = original
    spacy_tokens = spacy_lemmas.split()
    nltk_tokens = nltk_lemmas.split()
    max_len = max(len(original_tokens), len(spacy_tokens), len(nltk_tokens))

    for i in range(max_len):
        orig_token = original_tokens[i] if i < len(original_tokens) else ""
        spacy_token = spacy_tokens[i] if i < len(spacy_tokens) else ""
        nltk_token = nltk_tokens[i] if i < len(nltk_tokens) else ""
        print(f"{orig_token:<20} | {spacy_token:<20} | {nltk_token:<20}")

print_side_by_side(tokens, spacy_lemmatized_text, nltk_lemmatized_text)
```

Both does a good job removing the plural form, past tense to present tense, and removing 'ing'.
